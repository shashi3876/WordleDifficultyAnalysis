```python
# Importing the required libraries
import requests
import pandas as pd
from bs4 import BeautifulSoup
url = "https://wordfinder.yourdictionary.com/wordle/answers/"
data = requests.get(url).text

# Creating BeautifulSoup object
soup = BeautifulSoup(data, 'html.parser')
```

```python
print('Classes of each table:')
tables = soup.find_all('table')
df = pd.DataFrame(columns=['Date', 'Wordle Puzzle', 'Wordle Answer'])
print(len(tables))
# Collecting Ddata
for table in tables:
  print(table.tbody)
  for row in table.tbody.find_all('tr'):    
      # Find all data for each column
      columns = row.find_all('td')     
      #print(columns[0].text.strip())
      if(columns != []):
          date = columns[0].text.strip()
          number = columns[1].text.strip()
          WordleAnswer = columns[2].text.strip()       

          df = df.append({'Date': date,  'Wordle Puzzle': number, 'Wordle Answer': WordleAnswer}, ignore_index=True)
df.head(500)
```
